```python
import tabula

# Spécifiez le chemin du fichier PDF
pdf_path = "/content/sample_data/bna_efd311223.pdf"

# Utilisez l'option multiple_tables pour lire toutes les tables
dfs = tabula.read_pdf(pdf_path, pages='all', stream=True, multiple_tables=True)

# Affichez les tables
for df in dfs:
    print(df)
```

```python
import tabula
import pandas as pd

# Spécifiez le chemin du fichier PDF
pdf_path = "/content/sample_data/bna_efd311223.pdf"

# Utilisez l'option stream pour extraire toutes les données
dfs = tabula.read_pdf(pdf_path, pages='all', stream=True)

# Concaténez toutes les DataFrames extraites
result_df = pd.concat(dfs, ignore_index=True)

# Affichez les 5 premières lignes du DataFrame résultant pour vérifier
print(result_df.head())
```

# Convert PDF tables into CSV, TSV, or JSON files



```python
# You can convert from pdf into JSON, CSV, TSV

tabula.convert_into(pdf_path, "test.json", output_format="json", stream=True)
!cat test.json
```

```python
tabula.convert_into(pdf_path, "test.tsv", output_format="tsv", stream=True)
!cat test.tsv
```

```python
tabula.convert_into(pdf_path, "test.csv", output_format="csv", stream=True)
!cat test.csv
```
