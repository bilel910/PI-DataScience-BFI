```python
#discarded
import os
import cv2
import pytesseract
import unidecode

# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to the directory containing images
input_directory = r'D:\OutputNew'

# Path to the directory where filtered images will be saved
output_directory = r'D:\data_images'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Path to the directory where discarded images will be saved
discarded_directory = r'D:\ImageClassificationPI\AutresImages'

# Create the discarded images directory if it doesn't exist
os.makedirs(discarded_directory, exist_ok=True)

def contains_exclude_phrase(text):
    return ('etat des engagements hors bilan' in text.lower()) or ('introduction' in text.lower()) or ('opinion' in text.lower()) or ('afferents' in text.lower()) or ('referenciel' in text.lower())

def word_before_bilan(text):
    lines = text.split('\n')
    for line in lines:
        words = line.split()
        for word in words:
            if word in ['bilan', 'etat']:
                index = words.index(word)
                if index > 0:
                    return True
    return False

# Iterate over all directories (folders) in the input directory
for root, dirs, files in os.walk(input_directory):
    for filename in files:
        # Check if the file is an image (you can add more image extensions if needed)
        if (filename.endswith('.jpg') or filename.endswith('.png')):
            try:
                # Extract the number from the filename
                number = filename.split('_')[-1].split('.')[0]  # Assuming the number is after the last underscore and before the extension
                if len(number) <= 2 and 0 <= int(number) <= 7:
                    # Construct the full path to the input image
                    input_filepath = os.path.join(root, filename)
                    
                    # Read the image
                    image = cv2.imread(input_filepath)

                    # Convert image to grayscale
                    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

                    # Apply Gaussian blur for noise reduction
                    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

                    # Apply adaptive thresholding to enhance text visibility
                    thresh_image = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

                    # Use Pytesseract to perform OCR
                    text = pytesseract.image_to_string(thresh_image)
                    
                    # Remove accents from the text
                    text = unidecode.unidecode(text)

                    # Check if the image contains the exclude phrase
                    if contains_exclude_phrase(text.lower()):
                        # Construct the output file path for discarded images
                        folder_name = os.path.basename(root)
                        output_filename = f"{folder_name}_{filename}"  # Add folder name as prefix
                        discarded_output_filepath = os.path.join(discarded_directory, output_filename)

                        # Save the image to the discarded images directory
                        cv2.imwrite(discarded_output_filepath, image)
                        print(f"Image '{filename}' discarded and saved to '{discarded_output_filepath}'.")
                        continue

                    # Check if the word "bilan" is in the first 5 words of the text
                    if ('bilan' in text.lower()) or ('etat de resultat' in text.lower()) or ('etat des flux de tresorerie' in text.lower()) or ('etat de flux de tresorerie' in text.lower()) or ('flux' in text.lower()):
                        if word_before_bilan(text.lower()) == False:
                            # Construct the output file path
                            folder_name = os.path.basename(root)
                            output_filename = f"{folder_name}_{filename}"  # Add folder name as prefix
                            output_filepath = os.path.join(output_directory, output_filename)

                            # Save the image to the output directory
                            cv2.imwrite(output_filepath, image)
                            print(f"Image '{filename}' kept as it contains 'word' and saved to '{output_filepath}'.")
                    else:
                        # Construct the output file path for discarded images
                        folder_name = os.path.basename(root)
                        output_filename = f"{folder_name}_{filename}"  # Add folder name as prefix
                        discarded_output_filepath = os.path.join(discarded_directory, output_filename)

                        # Save the image to the discarded images directory
                        cv2.imwrite(discarded_output_filepath, image)
                        print(f"Image '{filename}' discarded as it does not contain 'word' and saved to '{discarded_output_filepath}'.")
                else:
                    # Construct the output file path for discarded images
                    folder_name = os.path.basename(root)
                    output_filename = f"{folder_name}_{filename}"  # Add folder name as prefix
                    discarded_output_filepath = os.path.join(discarded_directory, output_filename)

                    # Save the image to the discarded images directory
                    cv2.imwrite(discarded_output_filepath, image)
                    print(f"Image '{filename}' discarded as the number after the underscore is not between 0 and 6 or has more than two digits and saved to '{discarded_output_filepath}'.")
            except ValueError:
                print(f"Image '{filename}' discarded as it does not have a valid number after the underscore.")
```

```python
import os
import cv2
import pytesseract
import unidecode
import json
import re

# Read the image
image = cv2.imread(r'D:\output\aetech_efi300623\page_2.png')

# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur for noise reduction
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Apply adaptive thresholding to enhance text visibility
thresh_image = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Use Pytesseract to perform OCR
text = pytesseract.image_to_string(thresh_image)

# Remove accents from the text
text = unidecode.unidecode(text)

# Split the text into lines
lines = text.split("\n")

# Create a pattern to find decimal numbers with or without commas separated by spaces
pattern = r"\d{1,3}(?:[.,]\d{3})*(?:,\d+)?(?:\s+\d{1,3}(?:[.,]\d{3})*(?:,\d+)?)"

# Create a dictionary to store the extracted data
extracted_data = {}

# Process each line to extract table fields
for line in lines:
    # Find all matches of the pattern in the line
    matches = re.findall(pattern, line)
    if matches:
        # Extract the text before the first match as the key
        key = line.split(matches[0])[0].strip()
        # The matches are considered as values
        values = [match.strip() for match in matches]
        extracted_data[key] = values

# Convert the dictionary to JSON
json_data = json.dumps(extracted_data, ensure_ascii=False, indent=4)

# Print or save the JSON data
print(json_data)
# with open("output.json", "w", encoding="utf-8") as json_file:
#     json_file.write(json_data)
```
